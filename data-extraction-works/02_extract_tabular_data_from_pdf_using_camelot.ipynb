```python
import camelot
import pandas as pd
```

```python
tables = camelot.read_pdf("redacted1.pdf", flavor="stream", pages="all")
```

```python
len(tables)
```

```python
tables[3].df
```

```python
tables[1].df.to_csv('foo1.csv')
```

```python
tables.export('exam_one_raw.csv', f='csv', compress=True)
```

```python
data_df = []
for idx in range(len(tables)):
  data_df.append(tables[idx].df)
```

```python
len(data_df)
```

```python
all_data_df = pd.concat(data_df)
```

```python
all_data_df.reset_index(drop=False, inplace=True)
```

```python
# export to csv file
all_data_df.to_csv("exam_one_raw.csv")
```

```python
# save to json file
all_data_df.to_json('test1.json', orient="columns")
```
